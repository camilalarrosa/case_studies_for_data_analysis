```python
# Libs Necessárias

# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para anaálises gráficas
import matplotlib.pyplot as plt
import seaborn as sns

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
# Lendo os dados
Base_Dados = pd.read_csv('/content/sample_data/Salary_Data.csv')

# Verificando
Base_Dados.head()
```

```python
# Renomenado as colunas
Base_Dados.rename( columns={
    'YearsExperience' : 'Xp',
    'Salary' : 'Renda'
}, inplace=True )
```

```python
# dimensao
Base_Dados.shape
```

```python
# Campos nulos
Base_Dados.isnull().sum()
```

```python
# Grafico Nulos
sns.heatmap( Base_Dados.isnull(), cbar=False );
```

```python
Base_Dados.describe()
```

```python
# grafico distribuição para renda
plt.figure( figsize=(10, 6) )
plt.title('Análise da Renda')
sns.kdeplot( Base_Dados['Renda'], color='green' );
```

```python
# grafico distribuição para experiência
plt.figure( figsize=(10, 6) )
plt.title('Análise da Xp')
sns.kdeplot( Base_Dados['Xp'], color='blue' );
```

```python
# Distribuição por boxplot da renda
plt.figure( figsize=(10, 5) )
sns.boxplot( Base_Dados['Renda'] );
```

```python
# Distribuição por boxplot da experiência
plt.figure( figsize=(10, 5) )
sns.boxplot( Base_Dados['Xp'] );
```

```python
# scatterplot --> será que tem alguma relação linear?!
plt.figure( figsize=(10, 5) )
sns.scatterplot( data=Base_Dados, x='Renda', y='Xp');
```

```python
#comprovando a correlação
plt.figure( figsize=(10, 5) )
sns.regplot( data=Base_Dados, x='Renda', y='Xp');
```

```python
# Correlação
Correlacao = Base_Dados.corr('pearson')
plt.figure( figsize=(10, 5) )
sns.heatmap( Correlacao, annot=True, cbar=True );
```

```python
# Correlação
Correlacao = Base_Dados.corr('spearman')
plt.figure( figsize=(10, 5) )
sns.heatmap( Correlacao, annot=True, cbar=True );
```
